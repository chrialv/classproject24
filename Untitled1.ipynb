```python
def job_descriptions(ID):
  switcher = {
      "100": "Job Description: Software Engineer",
      "200": "Job Description: Technology Lawyer",
      "300": "Job Description: Graphics Designer"
  }

  return switcher.get(ID, "No Job Decription Match Found")

job_id = input("Enter a Job ID: ")
print(job_descriptions(job_id))

```

```python
from logging import info
import requests, re
from urllib.request import urlopen

def get_ip():
  url_in = "http://checkip.dyndns.org"

  response = requests.get(url_in, verify = True)
  if response.status_code != 200:
    return f"Status: {response.status_code}. Issue with request. Script will close."
    exit()

    info = str(urlopen(url_in).read())
    data = re.compile(r'Address: (\d+\.\d+\.\d+\.\d+\)').search(info).group(1)

    return data

your_ip = get_ip()
print(f"Your public IP address is: {your_ip}")
```

```python
import psutil

print("---- CPU Information ----")
print(f"CPUs Available: {psutil.cpu_count()}")
cpu_pre = psutil.cpu_percent(interval=1)
log_cores = psutil.cpu_count(logical=True)
print(f"Current CPU Usage: {cpu_pre}")
print(f"Number of logical CPU Cores {log_cores}")
phy_cores = psutil.cpu_count(logical=False)
print(f"Number of Physical CPU Cores: {phy_cores}")

print("---- Memory Information ----")
memory_details = psutil.virtual_memory()
print(f"Total Memory Available: {memory_details.total / (1024.0 ** 3):.2f}Gb")
print(f"Total Memory Utilised: {memory_details.used/ (1024.0 ** 3):2f}Gb")
print(f"Total Memory Remaining: {memory_details.available / (1024.0 ** 3):.2f}Gb")

print(" ---- Disk Information ----")
disk_usage = psutil.disk_usage('/')
print(f"Total Disk Space: {disk_usage.total / (1024.0 ** 3):.2f}GB")
print(f"Used Disk Space: {disk_usage.used / (1024.0 ** 3):.2f}Gb")
print(f"Free Disk Space: {disk_usage.free / (1024.0 ** 3):2f}Gb")
```

```python
import requests
from bs4 import BeautifulSoup

url_to_scrape = "https://tafensw.edu.au"
url_details = requests.get(url_to_scrape)
page_content = url_details.content

content_souped = BeautifulSoup(page_content, 'html.parser')

found_p_tags = content_souped.find_all('p')
for p_tag in found_p_tags:
  print(p_tag.text)
```

```python
from spellChecker import spellChecker
spell_list = spellChecker()

word_list = ['Pencil Sharner', 'Lettuce', 'Sciense', 'Keys']

for check_word in word_list:
  bad_word = check_word.lower()

  if bad_word in spell_list:
    print(f"{bad_word} is spelt correctly")
  else:
    print(f"{bad_word} is spelt incorrectly")
    closest = spell_list.correction(bad_word)
    print(f"The best suggestion for for {bad_word} is {closest}")
```
